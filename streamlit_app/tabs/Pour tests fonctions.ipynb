```python
import umap
import numpy as np
import matplotlib.pyplot as plt

# Assuming your dataframe is named df1
data = select_percentile_df.iloc[:, :-1].values  # Extracting all columns except the last one

# Extract the class/category column
classes = select_percentile_df.iloc[:, -1].values

# Create a UMAP object
umap_model = umap.UMAP()

# Fit the UMAP model to the data
umap_result = umap_model.fit_transform(data)

# Visualize the results
plt.scatter(umap_result[:, 0], umap_result[:, 1], c=classes)
plt.title('UMAP Projection')
plt.xlabel('UMAP Dimension 1')
plt.ylabel('UMAP Dimension 2')
plt.show()
```
